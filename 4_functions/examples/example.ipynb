Temperature Conversion

```python
def celtofeh(degree,unit):
    if unit == 'C':
        return degree * 9/5 +32
    elif unit == 'F':
        return (degree - 32) *5/9
    else:
        return None
    
print(celtofeh(77,'F'))
celtofeh(25,'C')
```

Password strength Checker

```python
def isStrongPass(password):
    if len(password) < 8:
        return False
    if not any(char.isdigit() for char in password):
        return False
    if not any(char.islower() for char in password):
        return False
    if not any(char.isupper() for char in password):
        return False
    if not any(char in '!@#$%^&*()+_-=' for char in password):
        return False
    return True

print(isStrongPass('weaKpass'))
print(isStrongPass('Pass@123'))
```

Total cost of items in cart

```python
def total(cart):
    totalCost = 0
    for item in cart:
        totalCost+=item['price']*item['quantity']
    return totalCost

cart = [
    {'item': 'apple','price':0.5,'quantity':4},
    {'item': 'banana','price':0.3,'quantity':8},
    {'item': 'cherry','price':0.7,'quantity':3}
]

bill = total(cart)
print(bill)
```

Check IF String is Palindrome

```python
def isPalindrome(string):
    if string[::] == string [::-1]:
        print("is Palindrome")
    else:
        print("Not a Palindrome")

check = isPalindrome('vidit')
```

```python
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n-1)

print(factorial(5))
```

Word Frequency Check from other files

```python
def freqCheck(filePath):
    wordCount ={}
    with open(filePath,'r') as file:
        for line in file:
            words = line.split()
            for word in words:
                word.lower().strip('.')
                wordCount[word] = wordCount.get(word,0)+1

    return wordCount

filepath='test.txt'
wordFreq = freqCheck(filepath)
print(wordFreq)
```
