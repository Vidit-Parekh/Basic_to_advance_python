```python
import sqlite3 as sql
```

```python
def create_database():
    conn = sql.connect('Employees.db')
    print('Creation Successfull')
    conn.close()

create_database()
```

```python
def create_Table():
    conn = sql.connect('Employees.db')
    cursor = conn.cursor()
    cursor.execute('''
CREATE TABLE IF NOT EXISTS Employees(
                   id INTEGER PRIMARY KEY,
                   name TEXT NOT NULL,
                   age INTEGER,
                   department TEXT
                   )
                   
''')
    conn.commit()
    print('Table Created!!')
    conn.close()

create_Table()
```

```python
def Insert_Data(id,name,age,department):
    conn = sql.connect('Employees.db')
    cursor = conn.cursor()
    cursor.execute('''
INSERT INTO EMPLOYEES(id,name,age,department)
                   VALUES (?,?,?,?)
''',(id,name,age,department))
    print('Data Inserted')
    conn.commit()
    conn.close()

Insert_Data(1,'Vidit',22,'IT')
    
```

```python
def Fetch_Data():
    conn = sql.connect('Employees.db')
    cursor = conn.cursor()
    cursor.execute('''
SELECT * FROM EMPLOYEES
''')
    data = cursor.fetchall()
    for data in data:
        print(data)
    conn.close()


Fetch_Data()
```

```python
def Fetch_Selective_Data(department):
    conn = sql.connect('Employees.db')
    cursor = conn.cursor()
    cursor.execute('''
SELECT * FROM EMPLOYEES WHERE department = ?
''',(department,))
    data = cursor.fetchall()
    for data in data:
        print(data)
    conn.close()

Fetch_Selective_Data('IT') 
```

```python
def update_Data(id,department):
    conn = sql.connect('Employees.db')
    cursor = conn.cursor()
    cursor.execute('''
UPDATE Employees
                   SET department = ?
                   WHERE id = ?

''',(department,id))
    conn.commit()
    conn.close()

update_Data(3,'Devops')
update_Data(2,'R&D')
update_Data(4,'Admin')
```

```python
def age_Specific(age):
    conn = sql.connect('Employees.db')
    cursor = conn.cursor()
    result = cursor.execute('SELECT * FROM Employees WHERE age < ?',(age,))
    for i in result:
        print(i)
    conn.commit()
    conn.close()

age_Specific(30)
```

```python
def Insert_multiple(data):
    con = sql.connect('Employees.db')
    cursor = con.cursor()
    try:
        cursor.execute('INSERT INTO Employees(id,name,age,department) VALUES (?,?,?,?)',data)
        con.commit()
    except Exception as e:
        con.rollback()
        print('Opps! error occured Transaction Rolled Back!!!')
        print('Error:',e)
    finally:
        con.close()

    
data = [[6, 'Frank', 40, 'Finance'],
    [7, 'Grace', 29, 'Engineering'],
    [8, 'Hannah', 35, 'Marketing'],
    [9, 'Ivan', 38, 'Sales'],
    ]

Insert_multiple(data)
```

```python
def update_Multiple_age(records):
    con = sql.connect('Employees.db')
    cursor = con.cursor()
    try:
        cursor.execute('UPDATE Employees SET age = ? WHERE id = ?',(records))
        con.commit()
    except Exception as e:
        con.rollback()
        print(e)
    finally:
        con.close()

records = [
    (32, 1),
    (26, 2),
    (33, 3),
    (41, 4),  # Non-existing ID to cause an error
    (23, 5)]
update_Multiple_age(records=records)
```

```python
def create_departments_table():
    conn = sql.connect('Employees.db')
    cursor = conn.cursor()
    cursor.execute('''
        CREATE TABLE IF NOT EXISTS departments (
            id INTEGER PRIMARY KEY,
            name TEXT NOT NULL
        )
    ''')
    conn.commit()
    conn.close()
    print("Table 'departments' created successfully.")

# Test the function
create_departments_table()
```
