Sampling

```python
import numpy as np
from sklearn.datasets import make_classification
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler
from collections import Counter
```

```python
# creation of imbalanced data
x,y = make_classification(n_classes=2,class_sep=2,weights=[0.1,0.9],n_informative=3,n_redundant=1,flip_y=0,n_features=20,n_clusters_per_class=1,n_samples=1000,random_state=42)

print(Counter(y))
```

```python
#oversample

oversample = RandomOverSampler(sampling_strategy='minority')
x_over,y_over = oversample.fit_resample(x,y)
print(Counter(y_over))
```

```python
#undersample

undersample = RandomUnderSampler(sampling_strategy='majority')
x_under,y_under = undersample.fit_resample(x,y)
print(Counter(y_under))
```

Balancing Techniques

```python
from sklearn.ensemble import RandomForestClassifier
from imblearn.ensemble import BalancedBaggingClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report 
```

```python
xtrain, xtest,ytrain,ytest = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
clasifier = RandomForestClassifier(random_state=42)

bb = BalancedBaggingClassifier(clasifier,sampling_strategy='auto',replacement=False,random_state=42)

```

```python
bb.fit(xtrain,ytrain)
```

```python
ypred = bb.predict(xtest)
```

```python
print(accuracy_score(ypred,ytest))
```

```python
print(classification_report(ypred,ytest))
```

Using SMOTE

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from collections import Counter
```

```python
x, y = make_classification(n_classes=2, class_sep=2, weights=[0.1, 0.9],
                           n_informative=3, n_redundant=1, flip_y=0,
                           n_features=20, n_clusters_per_class=1,
                           n_samples=1000, random_state=42)
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(Counter(ytrain))
```

```python
smote = SMOTE(sampling_strategy='auto',random_state=42)
```

```python
rx,ry = smote.fit_resample(xtrain,ytrain)
```

```python
print(Counter(ry))
```
