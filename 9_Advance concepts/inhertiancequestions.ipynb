```python
class Animal:
    def __init__(self,name,species):
        self.name = name
        self.species = species

    
class Dog(Animal):
    def __init__(self, name, species,breed):
        super().__init__(name, species)
        self.breed = breed
    def get_details(self):
        print(f"Name: {self.name} \nSpecies: {self.species} \nBreed: {self.breed}")

dog = Dog('Max','Dog','Husky')
dog.get_details()        
```

```python
class Animal:
    def __init__(self,name,species):
        self.name = name
        self.species = species

    
class Dog(Animal):
    def __init__(self, name, species,breed):
        super().__init__(name, species)
        self.breed = breed
    def __str__(self):
        return f"Name: {self.name} \nSpecies: {self.species} \nBreed: {self.breed}"
    def bark(self):
        return "BOUA-wow"

dog = Dog('Max','Dog','Husky')
print(dog)
print(dog.bark())      
```

```python
class Walker:
    def walk(self):
        print("Walking...")

class Runner:
    def run(self):
        print("Running...")

class Athlete(Walker, Runner):
    def __init__(self, training_hours):
        self.training_hours = training_hours

    def train(self):
        print(f"Training for {self.training_hours} hours.")

    def walk(self):
        print("Athlete walking...")
        super().walk()

athlete = Athlete(6)
athlete.walk()
athlete.train()
```

```python
class A:
    def show(self):
        print('Message in A')

class B(A):
    def show(self):
        print('Message in B')

class C(A):
    def show(self):
        print('Message in C')
    
class D(B,C):
    pass

d = D()
d.show()
```

```python
class Shape:
    def __init__(self,color):
        self.color = color

class Circle(Shape):
    def __init__(self,color,radius):
        super().__init__(color)
        self.radius = radius

c = Circle('blue',5)

print(f"the color of the circle with radius {c.radius} is {c.color}")
```

```python
class Person:
    def __init__(self,name):
        self.name = name

class Employee:
    def __init__(self,emp_id):
        self.emp_id = emp_id

class Manager(Person,Employee):
    def __init__(self, name,emp_id):
        super().__init__(name)
        Employee.__init__(self,emp_id)

m = Manager('Vidit',1224455)
print(f"the employee id {m.emp_id} holds {m.name}")
```

```python
class Vehicle:
    def start(self):
        print('Vehile Starting!!!') 
    
class Car(Vehicle):
    def start(self):
        print('Car Starting!!!')
        super().start()

car = Car()
car.start()
```

```python
class Base1:
    def __init__(self,a):
        self.a = a 

class Base2:
    def __init__(self,b):
        self.b = b 

class Base3(Base1,Base2):
    def __init__(self, a,b,c):
        super().__init__(a)
        Base2.__init__(self,b)
        self.c = c

base = Base3(2,3,4)

print(f"a:{base.a},b:{base.b},c:{base.c}")
```

```python
class Animal:
    pass

class Cat(Animal):
    pass


animal = Animal()
cat = Cat()
print(isinstance(animal, Animal))  # True
print(isinstance(cat, Animal))  # True
print(isinstance(cat, Cat))  # True
print(isinstance(animal, Cat))  # False
```

```python
class Bird:
    def Speak(self):
        pass
    
class Parrot(Bird):
    def Speak(self):
        print('Parrot Speaking!!')
    
class Penguin(Bird):
    def Speak(self):
        print('Penguin Speaking!!')
    
Birds = [Parrot(),Penguin()]
for bird in Birds:
    bird.Speak()

```

```python
class Device:
    def __init__(self,brand):
        self.brand = brand

class Phone(Device):
    def __init__(self,brand,model):
        super().__init__(brand)
        self.model = model

class Camera:
    def __init__(self,resolution):
        self.resolution = resolution

class Smartphone(Phone,Camera):
    def __init__(self, brand, model,resolution):
        Phone.__init__(self, brand, model)
        Camera.__init__(self, resolution)

s = Smartphone('Apple','16 PRO MAX','128MP X 3')
print(f"{s.brand} {s.model} {s.resolution}")
```
